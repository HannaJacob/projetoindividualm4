## üîπ 1. Importa√ß√£o de Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Nesta etapa, importamos as bibliotecas necess√°rias, incluindo pandas para manipula√ß√£o de dados e matplotlib.pyplot para visualiza√ß√£o.

## üîπ 2. Dados Fornecidos

```python
dados = {
    'Dia': ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta', 'S√°bado', 'Domingo'],
    'Horas_Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs_Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas_Concluidas': [5, 4, 6, 4, 5, 3, 2]
}
```

Definimos os dados fornecidos em um dicion√°rio, onde as chaves representam os nomes das colunas e os valores s√£o listas correspondentes.

## üîπ 3. Criar DataFrame

```python
df = pd.DataFrame(dados)
```

Criamos um DataFrame usando a biblioteca pandas para facilitar a manipula√ß√£o dos dados.

## üîπ 4. Adicionar Coluna de Produtividade Di√°ria

```python
df['Produtividade_Diaria'] = df['Tarefas_Concluidas'] / df['Horas_Trabalhadas']
```

Adicionamos uma nova coluna chamada 'Produtividade_Diaria', que representa o n√∫mero de tarefas conclu√≠das por hora.

## üîπ 5. Calcular M√©tricas

```python
total_horas_trabalhadas = df['Horas_Trabalhadas'].sum()
media_diaria_horas_trabalhadas = df['Horas_Trabalhadas'].mean()

total_bugs_corrigidos = df['Bugs_Corrigidos'].sum()
media_diaria_bugs_corrigidos = df['Bugs_Corrigidos'].mean()

total_tarefas_concluidas = df['Tarefas_Concluidas'].sum()
media_diaria_tarefas_concluidas = df['Tarefas_Concluidas'].mean()
```

Calculamos v√°rias m√©tricas, como total e m√©dia di√°ria de horas trabalhadas, bugs corrigidos e tarefas conclu√≠das.

## üîπ 6. Imprimir Resultados

```python
print(f'Total de Horas Trabalhadas: {total_horas_trabalhadas} horas')
print(f'M√©dia Di√°ria de Horas Trabalhadas: {media_diaria_horas_trabalhadas:.2f} horas')

print(f'Total de Bugs Corrigidos: {total_bugs_corrigidos} bugs')
print(f'M√©dia Di√°ria de Bugs Corrigidos: {media_diaria_bugs_corrigidos:.2f} bugs')

print(f'Total de Tarefas Conclu√≠das: {total_tarefas_concluidas} tarefas')
print(f'M√©dia Di√°ria de Tarefas Conclu√≠das: {media_diaria_tarefas_concluidas:.2f} tarefas')
```

Imprimimos os resultados das m√©tricas calculadas para an√°lise.

## üîπ 7. Visualiza√ß√£o

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(df['Dia'], df['Produtividade_Diaria'], marker='o', label='Produtividade Di√°ria')
ax.set_title('Produtividade Di√°ria (Tarefas Conclu√≠das por Hora)')
ax.set_xlabel('Dia')
ax.set_ylabel('Tarefas Conclu√≠das por Hora')
ax.legend()
plt.show()
```

Criamos um gr√°fico de linha para visualizar a produtividade di√°ria, facilitando a interpreta√ß√£o dos resultados.
